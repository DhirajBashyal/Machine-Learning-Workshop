**Text Preprocessing**

* **Tokenization :** Split text into smaller parts (words or sentences)
* **Stopwords :** Remove common words that don’t add meaning
* **Lemmatization:** Convert words to their base/root form
* **Cleaning text: ** Remove unwanted things: punctuation, numbers, special characters

```python
# Lemmatization
import spacy

nlp = spacy.load("en_core_web_sm")

tokens = ["running", "runs", "ran"]

# Convert list to text
doc = nlp(" ".join(tokens))

lemmas = [token.lemma_ for token in doc]

print(lemmas)
```

Word embeddings are a type of word representation in Natural Language Processing (NLP) that converts text into dense numerical vectors, capturing semantic meaning and contextual relationships.

```python
from gensim.models import Word2Vec

sentences = [
    ["i", "love", "nlp"],
    ["nlp", "is", "fun"],
    ["i", "enjoy", "learning"]
]

model = Word2Vec(sentences, vector_size=50, window=2, min_count=1)

print(model.wv["nlp"])
```

```python
import nltk
nltk.download('punkt_tab')
from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize

text = """Lorem Ipsum is simply dummy text of the printing and typesetting industry. Lorem Ipsum has been the industry's standard dummy text ever since the 1500s."""

tokens = word_tokenize(text.lower())

sentences = [tokens]


model = Word2Vec(
    sentences=sentences,
    vector_size=50,
    window=2,
    min_count=1,
    workers=4
)


print(model.wv["lorem"])
```

```python
# Sentiment Analysis : Positive / Negative classification
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

texts = ["I love this", "I hate this"]
labels = [1, 0]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

model = MultinomialNB()
model.fit(X, labels)

test = vectorizer.transform(["I hate it"])
print(model.predict(test))
```

**Named Entity Recognition**
Named Entity Recognition (NER) is a Natural Language Processing (NLP) technique that automatically identifies and classifies key information—entities—in unstructured text into predefined categories like names, organizations, locations, times, and monetary values

image.png


```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Elon Musk founded OpenAI in San Francisco."

doc = nlp(text)

for ent in doc.ents:
    print(ent.text, ent.label_)
```

**Text Classification**

* News classification
* Spam detection


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

texts = ["spam message", "important email"]
labels = [0, 1]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)

model = LogisticRegression()
model.fit(X, labels)

test = vectorizer.transform(["important message"])
print(model.predict(test))
```

**Stemming** is a technique used to extract the base form of the words by removing affixes from them. It is just like cutting down the branches of a tree to its stems. For example, the stem of the words eating, eats, eaten is eat.


**Lemmatization** technique is like stemming. The output we will get after lemmatization is called ‘lemma’, which is a root word rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word that means the same thing.
